```python
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt

#importing all libraries required

y0 = 1 #initial condition

start = 0
end = 2
t = np.linspace(start,end,end) #setting up the interval for t



def dYdt(y,t):
    #the differential equation 
    dydt = -2 * y * t
    return dydt

yinitial = np.linspace(y0,y0,end)


y = sc.integrate.odeint(dYdt,y0,t)

print('The answer for the ODE is: ', y)

#plotting the function
plt.figure()
plt.plot(t, y, 'r','change in y value')
plt.plot(t, yinitial, 'b--', linewidth=2)
plt.title('integration of dydt = -2yt')
plt.xlabel('Time')
plt.ylabel('y value')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```
